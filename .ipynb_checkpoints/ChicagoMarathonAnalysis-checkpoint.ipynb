## IO Functions

```python
def save_pkl(df, filename):
    with open('data/'+filename,'wb') as fobj:
        dill.dump(df,fobj)
```

```python
def load_pkl(filename):
    with open('data/'+filename, 'rb') as fobj:
        return dill.load(fobj)
```

## Scraping functions

```python
def remove_prefix(raw, pref):
    re_str = pref+'(.+)'
    compiled_regex = re.compile(re_str)
    return re.findall(compiled_regex, raw)[0]
```

```python
def split_name(namefield):
    compiled_regex = re.compile(r'(.+) \(([A-Z]+)\)')
    try:
        name_country = re.findall(compiled_regex, namefield)[0]
    except IndexError:
        name_country = (namefield, None)
    return name_country
```

```python
def parse_time(time_string):
    dt = parse(time_string)-parse('00:00:00')
    return dt.seconds/3600
```

```python
def get_fields(row):
    
    name_country = split_name(row.find_all('h4', class_=' list-field type-fullname')[0].text)
    try:
        data_row = [int(row.find_all('div', class_=' list-field type-place place-secondary hidden-xs numeric')[0].text), #overall place
                    int(row.find_all('div', class_=' list-field type-place place-primary numeric')[0].text), #gender place
                    name_country[0], #name
                    name_country[1], #country
                    remove_prefix(row.find_all('div', class_=' list-field type-field')[0].text, 'BIB'), #bib number
                    remove_prefix(row.find_all('div', class_=' list-field type-age_class')[0].text, 'Division')] # division
        try:
            half_time = parse_time(remove_prefix(row.find_all('div', class_='split list-field type-time')[0].text, 'HALF')) # half time
        except ValueError:
            half_time = None
            
        try:
            fin_time = parse_time(remove_prefix(row.find_all('div', class_=' list-field type-time')[0].text, 'Finish')) # finish time
        except ValueError:
            fin_time = None
            
        data_row.extend([half_time, fin_time])
                
        return data_row
    except IndexError:
        return None
```

## Scrape webpages

```python
# Set up variables
num_dnf = 0
sexes = ['M','W']
yrs = [str(yr) for yr in np.arange(2018,1995,-1)]
event_codes = ['MAR_999999107FA30900000000B5'#2018,
               'MAR_999999107FA30900000000A1', #2017
               'MAR_999999107FA309000000008D', #2016
               'MAR_999999107FA3090000000079', #2015
               'MAR_999999107FA3090000000065', #2014
               'MAR_9999990E9A92360000000079', #2013
               'MAR_9999990E9A9236000000003D', #2012
               'MAR_9999990E9A92360000000029', #2011
               'MAR_9999990E9A92360000000015', #2010
               'MAR_9999990E9A92360000000002', #2009
               'MAR_9999990E9A92360000000051', #2008
               'MAR_9999990E9A92360000000052', #2007
               'MAR_9999990E9A92360000000065', #2006
               'MAR_9999990E9A92360000000066', #2005
               'MAR_9999990E9A92360000000067', #2004
               'MAR_9999990E9A92360000000068', #2003
               'MAR_9999990E9A92360000000069', #2002
               'MAR_9999990E9A9236000000006A', #2001
               'MAR_9999990E9A9236000000006B', #2000
               'MAR_9999990E9A9236000000006C', #1999
               'MAR_9999990E9A9236000000006D', #1998
               'MAR_9999990E9A9236000000006E', #1997
               'MAR_9999990E9A9236000000006F' #1996
               ] 
```

```python
for count, event_code in enumerate(event_codes):
    yr_df = pd.DataFrame()
    for sex in sexes:
        data = []
        i = 1
        pg_found = True
        while pg_found:
            
            print(f'Page: {i}')
            result_pg = f'https://chicago-history.r.mikatiming.com/2018/?page={i}&event={event_code}&lang=EN_CAP&num_results=1000&pid=list&search%5Bsex%5D={sex}&search%5Bage_class%5D=%25'
            page = requests.get(result_pg)
            soup = BeautifulSoup(page.text, 'html.parser')
        
            rows = soup.find_all('li', class_=' list-active list-group-item row')
            rows.extend(soup.find_all('li', class_=' list-group-item row'))
            if len(rows) > 0:
                for row in rows:
                    data_row = get_fields(row)
                    if data_row:
                        data.append(data_row)
                    else:
                        num_dnf += 1
            else:
                pg_found = False
            i += 1
        sex_df = pd.DataFrame(data, columns=['Place', 'GenderPlace', 'Name', 'Country', 'BibNumber', 'Division', 'HalfTime', 'Time'])
        sex_df['Gender'] = [sex for i in range(len(sex_df))]
        if yr_df.empty:
            yr_df = sex_df            
        else:
            yr_df = pd.concat([yr_df, sex_df])
    filename = f'{yrs[count]}_results'
    save_pkl(yr_df, filename)
```

## Analysis functions

```python
def add_df_split(df):
    df['split_age'] = np.nan
    df.loc[(df.division=='16-19') | (df.division=='20-24') | (df.division=='25-29') | (df.division=='30-34') | (df.division=='35-39'),'split_age'] = '16-39'
    df.loc[(df.division!='16-19') & (df.division!='20-24') & (df.division!='25-29') & (df.division!='30-34') & (df.division!='35-39'),'split_age'] = '40+'

    return df
```

This shows that any effect on the numbers of qualifying times based on the BQ time adjustments is overshadowed by yearly variations. This is likely due to weather for any given race (2007 had record heat, and the course was even shut down early). In the section below I try to correct for this.

### Correcting for average time

```python
year_list = []
div_list = []
gender_list = []
num_qual_list = []
num_no_qual_list = []
prop_qual_list = []

for year in range(1996,2020):
    df = finishers_dict[str(year)]
    bq_df = bq_dict[str(year)]
    year_adjustment = df.Time.mean() / 4.5
    for count, division in enumerate(bq_divs):
        for sex in ['M', 'W']:
            bq_time = bq_df[bq_df['Division'] == division][sex].values[0]
            num_qual = df.loc[(df['Division']==division) & (df['Gender']==sex) & (df['Time']<=(bq_time*year_adjustment))].Time.count()
            num_no_qual = df.loc[(df['Division']==division) & (df['Gender']==sex) & (df['Time']>(bq_time*year_adjustment))].Time.count()
                                 
            year_list.append(year)
            div_list.append(division)
            gender_list.append(sex)
            num_qual_list.append(num_qual)
            num_no_qual_list.append(num_no_qual)
            prop_qual_list.append(num_qual/(num_qual+num_no_qual))
            
data = {'year':year_list, 'sex':gender_list, 'division':div_list, 'number_qualified':num_qual_list, 'number_not_qualified':num_no_qual_list, 'proportion_qualified':prop_qual_list}
data = pd.DataFrame(data)
save_pkl(data, 'bq_rates_adjusted')
            
```

```python
data_adj = load_pkl('bq_rates_adjusted')
```

```python
data_adj['split_age'] = np.nan
data_adj.loc[(data.division=='16-19') | (data.division=='20-24') | (data.division=='25-29') | (data.division=='30-34') | (data.division=='35-39'),'split_age'] = '16-39'
data_adj.loc[(data.division!='16-19') & (data.division!='20-24') & (data.division!='25-29') & (data.division!='30-34') & (data.division!='35-39'),'split_age'] = '40+'
```

```python
qualifying_number_adj = data_adj.groupby(['year','split_age', 'sex']).number_qualified.sum()
qualifying_proportion_adj = qualifying_number_adj / (qualifying_number_adj + data_adj.groupby(['year','split_age', 'sex']).number_not_qualified.sum())
```

```python

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16,4))
ax0, ax1 = axes.flatten()

qualifying_number_adj.unstack().unstack().plot(kind='line', ax=ax0)
qualifying_proportion_adj.unstack().unstack().plot(kind='line', ax=ax1)
for yr in [2001, 2011, 2018]:
    ax0.axvline(x=yr, ymax=2000, color=[0.4,0.4,0.4], linestyle='--')
    ax1.axvline(x=yr, color=[0.4,0.4,0.4], linestyle='--')
    
ax0.set_ylabel('Qualifying Number')
ax0.set_title('Chicago Marathon BQ-ers ("Weather" corrected)')
ax1.set_ylabel('Qualifying Proportion')
ax1.set_title('Chicago Marathon BQ-ers ("Weather" corrected)')

plt.show()
```

This correction for "weather" has limitations. For example, this suggests that runners were less likely to qualify for the Boston marathon pre-2000, but in fact the opposite is true when you look at the uncorrected numbers of qualifiers. This is likely indicative of two things. 1 - The qualifying standard was easier in the 90's, and 2 - In the last 2 decades marathons have become more mainstream, and thus the average marathon time has gone down.

Additionally, I still see lots of yearly variation. This is likely because the effect of heat on a runner in the middle of the pack, running relatively slower, has less effect than a runner who is closer to their peak performance.

To get a true idea of whether or not a BQ qualifying time adjustment affects I would fit raw times to a line or curve, and calculate deviations from the curve to determine if the deviation was significant. Although there are many runners in Chicago, the effect of a BQ time adjustment is likely masked by the yearly variation due to weather. To compensate for this, I would need to get data from more marathons.
